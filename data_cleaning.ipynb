## Airports Dataset

#### Import packages

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

##### Import dataset into dataframe


```python
airports_df = pd.read_csv('csv/airports.csv',encoding='latin-1')
```

#### Identify numeric and non-numeric columns

```python
numeric  = airports_df.select_dtypes(include=[np.number])
numeric.columns.values
```

```python
non_numeric = airports_df.select_dtypes(exclude=[np.number])
non_numeric.columns.values
```

#### Handle Missing Data

The dataset encodes missing values with \N, so these will be converted to the standard NaN format

```python
cols = airports_df.columns
for col in cols:
    airports_df.loc[airports_df[col] =='\\N',col] = np.NaN

#export the data
# airports_df.to_csv('csv/airports')
```

```python
colors = ['green', 'red']
sns.heatmap(airports_df[cols].isna(), cmap=sns.color_palette(colors))
```
